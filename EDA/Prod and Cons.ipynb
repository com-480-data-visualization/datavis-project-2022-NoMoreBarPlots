```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
```

```python
# Load data
Fuel_Oil = pd.read_csv('Fuel Oil.csv')
Fuel_Oil
```

```python
Fuel_Oil.groupby(['Transaction']).count()
```

```python
Prod = Fuel_Oil.loc[Fuel_Oil['Transaction']=='Production']
Prod

Cons = Fuel_Oil.loc[Fuel_Oil['Transaction']=='Final consumption']
Cons
```

```python
Prod_S = Prod.loc[Prod['Country or Area']=='Switzerland']
Prod_S

Cons_S = Cons.loc[Cons['Country or Area']=='Switzerland']
Cons_S
```

```python
plt.plot(Prod_S['Year'], Prod_S['Quantity'], label = 'Prod') 
plt.plot(Cons_S['Year'], Cons_S['Quantity'], label = 'Cons') 
plt.legend()
plt.title('Switzerland production and consumption of Fuel Oil')
```

```python
Prod_US = Prod.loc[Prod['Country or Area']=='United States']

Cons_US = Cons.loc[Cons['Country or Area']=='United States']
Cons_US
```

```python
plt.plot(Prod_US['Year'], Prod_US['Quantity'], label = 'Prod') 
plt.plot(Cons_US['Year'], Cons_US['Quantity'], label = 'Cons') 
plt.legend()
plt.title('US production and consumption of Fuel Oil')
```

```python
Prod_sum = []
Cons_sum = []
year = []
for y in range(min(Prod['Year']), max(Prod['Year'])) : 
    year.append(y)
    Prod_sum.append(sum(Prod['Quantity'].loc[Prod['Year']==y]))
    Cons_sum.append(sum(Cons['Quantity'].loc[Cons['Year']==y]))
```

```python
plt.plot(year, Prod_sum, label = 'Prod') 
plt.plot(year, Cons_sum, label = 'Cons') 
plt.legend()
plt.xlabel("Year")
plt.title('World production and consumption of Fuel Oil (sum over all contries)')
plt.savefig('Fuel_cons_prod.png')
```

```python
Elec1 = pd.read_csv('Total Electricity 1.csv')
Elec2 = pd.read_csv('Total Electricity 2.csv')

data = [Elec1, Elec2]
Elec = pd.concat(data)
Elec
```

```python
Elec.groupby(['Transaction']).count()
```

```python
Prod = Elec.loc[Elec['Transaction']=='net production']
Cons = Elec.loc[Elec['Transaction']=='Final energy consumption']

Prod_sum = []
Cons_sum = []
year = []
for y in range(min(Prod['Year']), max(Prod['Year'])) :
    year.append(y)
    Prod_sum.append(sum(Prod['Quantity'].loc[Prod['Year']==y]))
    Cons_sum.append(sum(Cons['Quantity'].loc[Cons['Year']==y]))
    
plt.plot(year, Prod_sum, label = 'Prod')
plt.plot(year, Cons_sum, label = 'Cons')
plt.legend()
plt.xlabel("Year")
plt.title('World production and consumption of Total Electricity (sum over all contries)')
plt.savefig('Elec_cons_prod.png')
```
